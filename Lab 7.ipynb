For Lab 7, we must compare different algorithms for effiency in compressing 1000 terabytes of data each day.  To start, we must first generate 100 megabytes of random binary data. ```python
import numpy as np

myvar = np.random.choice([0, 1], size = 102, replace = True, p = [0.5, 0.5])
print(myvar)
myvar = np.packbits(myvar)
open("myvar", "wb").write(myvar)

#print(myvar)
```

**Compressing real data:** My real data consists of 10 homologs of gp120 from different HIV isolates.  I will find these searching "gp120" in the ncbi nucleotide database.  In order to keep my seqeunces consistent, I chose to search for the whole gp120 gene and applied a search filter for sequence length (1300-1500 bps). ```python
import sys
from Bio import SeqIO

records=[]

seq1 = open("gp120seq1.fasta")
seq2 = open("gp120seq2.fasta")
seq3 = open("gp120seq3.fasta")
seq4 = open("gp120seq4.fasta")
seq5 = open("gp120seq5.fasta")
seq6 = open("gp120seq6.fasta")
seq7 = open("gp120seq7.fasta")
seq8 = open("gp120seq8.fasta")
seq9 = open("gp120seq9.fasta")
seq10 = open("gp120seq10.fasta")


records.append(SeqIO.read(seq1, "fasta"))
records.append(SeqIO.read(seq2, "fasta"))
records.append(SeqIO.read(seq3, "fasta"))
records.append(SeqIO.read(seq4, "fasta"))
records.append(SeqIO.read(seq5, "fasta"))
records.append(SeqIO.read(seq6, "fasta"))
records.append(SeqIO.read(seq7, "fasta"))
records.append(SeqIO.read(seq8, "fasta"))
records.append(SeqIO.read(seq9, "fasta"))
records.append(SeqIO.read(seq10, "fasta"))

SeqIO.write(records, "gp120seqs.fasta", "fasta")
```
